# 畫出地圖

```python
# forlium.initialize_notebook() is broken. (https://github.com/python-visualization/folium/issues/132)
# The following is a workaround.

def inline_map(m, width=650, height=500):
    """Takes a folium instance and embed HTML."""
    m._build_map()
    srcdoc = m.HTML.replace('"', '&quot;')
    embed = HTML('<iframe srcdoc="{}" style="width: {}px; height: {}px; border: none"></iframe>'.format(srcdoc, width, height))
    return embed

map_osm = folium.Map(location=[23.5, 121], zoom_start=7)
inline_map(map_osm)
```

```python
factorymap = folium.Map(location=[23.5, 121], zoom_start=7)
for r in range(0,df.shape[0]):
    info=u'<b>{name}</b><br/>網址: <a href="{url}">{url}</a><br/>簡介：<br/>{intro}'.encode("utf-8").format(name=df[u'工廠名稱'][r], url= df[u'網址'][r], intro=df[u'簡介'][r])
    factorymap.simple_marker([float(i) for i in df[u'經緯度'][r].strip('"').split(',')], popup=info)
inline_map(factorymap)
```
